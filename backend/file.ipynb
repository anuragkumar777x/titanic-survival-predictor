```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.drop(columns=['Cabin'], inplace=True)
```

```python
df.head()
```

```python
## put missing values for age
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df.info()
```

```python
## imputing missing values for embark

df['Embarked'].value_counts()

## S is most occuring 

df['Embarked'] = df['Embarked'].fillna('S', inplace=True)
```

```python
df.info()
```

```python
# Convert columns to appropriate data types

df['Survived'] = df['Survived'].astype('category')
df['Pclass'] = df['Pclass'].astype('category')
df['Sex'] = df['Sex'].astype('category')
df['Age'] = df['Age'].astype('int')
df['Embarked'] = df['Embarked'].astype('category')
```

```python
df.info()
```

```python
## five point summary 


df.describe()
```

```python
# Univariate Analysis - Survived Column

sns.countplot(x='Survived', data=df)

death_percent = round(
    (df['Survived'].value_counts()[0] / len(df)) * 100
)

print("Out of {}, {}% people died in accident".format(len(df), death_percent))
```

```python
# Pclass Column Analysis

print((df['Pclass'].value_counts(normalize=True)) * 100)

sns.countplot(x='Pclass', data=df)

# Pclass was the most crowded class
```

```python
print((df['Sex'].value_counts(normalize=True))*100)
sns.countplot(x='Sex', data=df)
```

```python
print((df['Embarked'].value_counts(normalize=True)) * 100)

sns.countplot(x='Embarked', data=df)
```

```python
sns.displot(df['Age'])

print(df['Age'].skew())
print(df['Age'].kurt())
```

```python
sns.boxplot(x='Age', data=df)
```

```python
print("Age between 60 and 70 : ", df[(df['Age'] >= 60) & (df['Age'] <= 70)].shape[0])
print("Age greater than 70 and up to 75 : ", df[(df['Age'] > 70) & (df['Age'] <= 75)].shape[0])

print("Age greater than 75 : ", df[df['Age'] > 75].shape[0])

print("-"*50)

print("Age between 0 and 1 (infants) : ", df[(df['Age'] >= 0) & (df['Age'] <= 1)].shape[0])
```
